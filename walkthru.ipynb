## Useful keyboard shortcuts for using notebooks

Press ESC (caps lock if using modified key) or Control (^) M to enter command mode

- Use a dash to create points

- Press Enter to enter edit mode
- In command mode press M to change to Markdown cell
- In edit mode press Y to change to Code cell

- Press Shift + Enter to move to next cell

- To break up a cell into separate rows
- This is what you do
- Place your cursor where you want the break and

- Press ^ + Shift + '-'

- Insert a cell above in command mode by pressing A
- Insert a cell below in command mode by pressing B
- Delete a cell in command mode by pressing D,D
